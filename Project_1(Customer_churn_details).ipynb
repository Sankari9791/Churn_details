```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.metrics import accuracy_score
```

```python
Customer_churn_data=pd.read_csv("Customer_churn_details.csv")
Customer_churn_data
```

```python
Customer_churn_data.columns
```

```python
drop_col=Customer_churn_data.drop(columns=['customerID', 'gender', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                                  'PaymentMethod', 'StreamingTV', 'StreamingMovies'],inplace=True)
     
```

```python
df_columns=Customer_churn_data.columns
Categorical_data=[x for x in Customer_churn_data if Customer_churn_data[x].dtype=='object']
print(Categorical_data)
Customer_churn_data.head()
```

```python
finding_null_data=Customer_churn_data.isna().sum()
print(finding_null_data)
```

```python
Categorical_mapping={"Yes":1,"No":0}
Customer_churn_data['Dependents']=Customer_churn_data['Dependents'].map(Categorical_mapping)
print(Customer_churn_data.info())
Customer_churn_data.head()
```

```python
Categorical_mapping={"Month-to-month":0,"One year":1,"Two year":2}
Customer_churn_data['Contract']=Customer_churn_data['Contract'].map(Categorical_mapping)
print(Customer_churn_data.info())
Customer_churn_data.head(20)                                    
                                        
                                           
```

```python

Categorical_mapping={"Yes":1,"No":0}
Customer_churn_data['PaperlessBilling']=Customer_churn_data['PaperlessBilling'].map(Categorical_mapping)
print(Customer_churn_data.info())
Customer_churn_data.head(20)
```

```python
Categorical_mapping={"Yes":1,"No":0}
Customer_churn_data['Partner']=Customer_churn_data['Partner'].map(Categorical_mapping)
print(Customer_churn_data.info())
Customer_churn_data.head(20)
```

```python
Categorical_mapping={"Yes":1,"No":0}
Customer_churn_data['Churn']=Customer_churn_data['Churn'].map(Categorical_mapping)
print(Customer_churn_data.info())
Customer_churn_data.head(20)
```

```python
# Make sure any leading/trailing whitespace is removed and empty strings are converted to NaN
Customer_churn_data['TotalCharges'] = Customer_churn_data['TotalCharges'].str.strip()

# Replace any empty strings with NaN before converting
Customer_churn_data['TotalCharges'] = Customer_churn_data['TotalCharges'].replace('', float('nan'))

# Convert the 'TotalCharges' column to numeric, coercing errors to NaN
Customer_churn_data['TotalCharges'] = pd.to_numeric(Customer_churn_data['TotalCharges'], errors='coerce')

# Display the DataFrame to check
print(Customer_churn_data[['TotalCharges']].head(20))
Customer_churn_data.info()
```

```python
missing_data=Customer_churn_data.isna().sum()
missing_data
```

```python
Customer_churn_data.fillna(0,inplace=True)
print("After Filling")
missing_data=Customer_churn_data.isna().sum()
missing_data
```

```python
Customer_churn_data.info()
```

```python
x=Customer_churn_data.drop(columns='Churn',axis=1)
x
```

```python
y=Customer_churn_data['Churn']
y
```

```python
#split the data for training and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#model fit
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predicted the model
x_trained_pred=model.predict(x_train)
accuracy=accuracy_score(x_trained_pred,y_train)
x_trained_pred
print(round(accuracy,2))
```

```python
x_test_pred=model.predict(x_test)
accuracy=accuracy_score(x_test_pred,y_test)
x_test_pred
print(round(accuracy,2))
```
